```python
from graphviz import Digraph

# Configurazione Stile
dot = Digraph(comment='Gradient Boosting', format='png')
dot.attr(rankdir='LR') # Left-to-Right
dot.attr('node', shape='box', style='filled', fillcolor='#E3F2FD', fontname='Sans-Serif')

# Input e Primo Modello
dot.node('Input', 'Input Data\n(x)', shape='oval', fillcolor='#CFD8DC')
dot.node('Tree1', 'Albero 1\n(Fit su y)', fillcolor='#C8E6C9')
dot.node('Pred1', 'Predizione 1\n(y_1)', shape='plaintext', fillcolor='white')

# Calcolo Errore
dot.node('Res1', 'Residuo 1\n(e1 = y - y_1)', shape='oval', fillcolor='#FFCCBC')

# Secondo Modello (corregge l'errore)
dot.node('Tree2', 'Albero 2\n(Fit su e1)', fillcolor='#C8E6C9')
dot.node('Pred2', 'Predizione 2\n(y_2)', shape='plaintext', fillcolor='white')

# Somma Finale
dot.node('Sum', 'Somma Additiva\nY_final = y_1 + y_2', shape='doubleoctagon', fillcolor='#FFF9C4')

# Collegamenti
dot.edge('Input', 'Tree1')
dot.edge('Tree1', 'Pred1')
dot.edge('Input', 'Res1', label='Target - Pred')
dot.edge('Pred1', 'Res1', style='dashed')
dot.edge('Res1', 'Tree2')
dot.edge('Tree2', 'Pred2')
dot.edge('Pred1', 'Sum')
dot.edge('Pred2', 'Sum')

# Render
dot.render('gradient_boosting_schema', view=False)
print("Schema generato: gradient_boosting_schema.png")
```
