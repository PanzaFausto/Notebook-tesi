```python
from graphviz import Digraph

# Creazione schema dello split
dot = Digraph(comment='Gini Split', format='png')
dot.attr(rankdir='TB') # Top-Bottom

# Nodo Padre
dot.node('A', 'Nodo T\n(Samples = 100)\nImpurity = 0.60', shape='box', style='filled', fillcolor='#E1F5FE')

# Decisione (Rombo)
dot.node('Decision', 'Beacon_26_1\n<= -85 dBm?', shape='diamond', style='filled', fillcolor='#FFF9C4')

# Nodi Figli
dot.node('B', 'Nodo Left (True)\n(Samples = 60)\nImpurity = 0.25', shape='box', style='filled', fillcolor='#C8E6C9')
dot.node('C', 'Nodo Right (False)\n(Samples = 40)\nImpurity = 0.75', shape='box', style='filled', fillcolor='#FFCCBC')

# Archi
dot.edge('A', 'Decision')
dot.edge('Decision', 'B', label='  Sì')
dot.edge('Decision', 'C', label='  No')

# Render
dot.render('gini_split_diagram', view=False)
print("Immagine gini_split_diagram.png generata!")
```
