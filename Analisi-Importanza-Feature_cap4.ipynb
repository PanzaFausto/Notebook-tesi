```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

# Ricarica dati Android
df_train = pd.read_csv('android_session_1_1_median.csv').fillna(-110)
beacon_cols = [c for c in df_train.columns if '-' in c and c not in ['room', 'artwork']]
X = df_train[beacon_cols]
y = df_train['artwork']

# Addestra
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Estrai Importanza
importances = rf.feature_importances_
feature_names = beacon_cols

# Crea DataFrame per il grafico
df_imp = pd.DataFrame({'Beacon': feature_names, 'Importance': importances})
df_imp = df_imp.sort_values(by='Importance', ascending=False).head(15) # Top 15 beacon

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Beacon', data=df_imp, palette='viridis')
plt.title('Top 15 Beacon più importanti per la classificazione (Random Forest)')
plt.xlabel('Mean Decrease Impurity (MDI)')
plt.ylabel('Beacon ID')
plt.grid(True, alpha=0.3)
plt.savefig('feature_importance.png', dpi=300)
print("Grafico feature_importance.png generato!")
```
