```python
from graphviz import Digraph

# Creazione diagramma
dot = Digraph(comment='Pipeline di Localizzazione', format='png')
dot.attr(rankdir='LR') # Da sinistra a destra

# Stile dei nodi
dot.attr('node', shape='box', style='filled', fillcolor='#E0F7FA', fontname='Sans-Serif')

# Input
dot.node('A', 'Raw Data\n(Android/iOS RSSI)', shape='parallelogram', fillcolor='#EEEEEE')

# Pre-processing
dot.node('B', 'Pre-processing\n(Time Window 3s\n& Median Filter)', fillcolor='#FFF9C4')

# Feature Engineering
dot.node('C', 'Feature Engineering\n(Imputation -110 dBm\n& Scaling)', fillcolor='#FFF9C4')

# Model Selection
dot.node('D', 'Model Selection\n(Grid Search &\n10-Fold CV)', fillcolor='#C8E6C9')

# Classificatori
dot.node('E', 'Final Model\n(Random Forest)', fillcolor='#FFCC80')

# Output
dot.node('F', 'Position Prediction\n(Room / Artwork)', shape='parallelogram', fillcolor='#EEEEEE')

# Collegamenti
dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D')
dot.edge('D', 'E')
dot.edge('E', 'F')

# Salva
dot.render('pipeline_schema', view=False)
print("Immagine pipeline_schema.png generata")
```
