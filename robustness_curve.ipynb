```python
import matplotlib.pyplot as plt

# Dati Ufficiali (estratti dalla Tabella 4.4)
# I veri step in cui ho registrato i guasti sono 5:
x_beacons_spenti = [0, 2, 4, 10, 20]

# 5 valori di Accuratezza per il Random Forest
y_rf = [89.77, 89.65, 89.88, 84.77, 81.70]

# 5 valori di Accuratezza per il k-NN
y_knn = [85.70, 85.23, 84.65, 81.73, 79.18]

# Configurazione stile grafico
plt.figure(figsize=(10, 6), dpi=300) # dpi=300 garantisce qualità tipografica

try:
    plt.style.use('seaborn-v0_8-whitegrid')
except:
    plt.style.use('seaborn-whitegrid') # Per versioni vecchie di matplotlib

# Plot delle due curve
# Random Forest: linea continua blu con pallini
plt.plot(x_beacons_spenti, y_rf, marker='o', linestyle='-', linewidth=2.5,
         markersize=8, label='Random Forest (Proposto)', color='#1f77b4')

# k-NN: linea tratteggiata rossa con quadrati
plt.plot(x_beacons_spenti, y_knn, marker='s', linestyle='--', linewidth=2.5,
         markersize=8, label='k-NN (Baseline)', color='#d62728')

# Personalizzazione Testi e Assi
plt.title('Stress Test Reale: Decadimento Prestazionale (Artwork Level)',
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Numero di Beacon Guasti', fontsize=12)
plt.ylabel('Accuratezza (%)', fontsize=12)

# Mostra sull'asse X ESATTAMENTE i punti in cui ci sono i dati
plt.xticks(x_beacons_spenti, fontsize=11)
plt.yticks(fontsize=11)

# Imposta i limiti dell'asse Y per inquadrare bene il calo senza appiattirlo
plt.ylim(75, 95)

# Legenda e Griglia
plt.legend(fontsize=11, loc='lower left')
plt.grid(True, linestyle=':', alpha=0.7)

# Salvataggio del file
nome_file = 'robustness_curve_REAL.png'
plt.savefig(nome_file, bbox_inches='tight')
print(f"Grafico generato e salvato con successo: {nome_file}")

# Mostra a schermo
plt.show()
```
