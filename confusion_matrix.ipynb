```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier

# Carica i dati
try:
    df_train = pd.read_csv('android_session_1_1_median.csv')
    df_test = pd.read_csv('android_session_2_2_median.csv')
except:
    print("File CSV non caricati")

# Prepara Feature e Target
beacon_cols = [c for c in df_train.columns if '-' in c and c not in ['room', 'artwork']]
X_train = df_train[beacon_cols].fillna(110)
y_train = df_train['room']
X_test = df_test[beacon_cols].fillna(110)
y_test = df_test['room']

# Addestra
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

# Genera Matrice con CONTEGGI (Numeri interi)
classes = sorted(y_test.unique())
cm = confusion_matrix(y_test, y_pred, labels=classes)

# Disegna
plt.figure(figsize=(12, 10))
# fmt='d' serve per stampare interi (digit), cmap='OrRd' (Orange-Red) evidenzia meglio gli errori
sns.heatmap(cm, annot=True, fmt='d', cmap='OrRd', cbar=False,
            xticklabels=classes, yticklabels=classes)
plt.ylabel('Vera Stanza (Ground Truth)')
plt.xlabel('Stanza Predetta')
plt.title(f'Matrice di Confusione (Errori Assoluti su {len(y_test)} campioni)')
plt.tight_layout()
plt.savefig('confusion_matrix_heatmap_COUNT.png', dpi=300)
plt.show()
```
